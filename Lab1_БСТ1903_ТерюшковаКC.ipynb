### Сортировка выбором

```python
def find_smallest(arr):
    smallest=arr[0]
    smallest_index=0
    for i in range(1, len(arr)):
        if arr[i]<smallest:
            smallest=arr[i]
            smallest_index=i
    return smallest_index
def sel_sort(arr):
    newArr=[]
    for i in range(len(arr)):
        smallest=find_smallest(arr)
        newArr.append(arr.pop(smallest))
    return newArr
array = sel_sort([x]) 
print (array)
```

### Сортировка вставкой


```python
def insertion_sort(array):

    for i in range(len(array)):
        n= array[i]  
        while array[i - 1] > n and i > 0:
            array[i] = array[i-1]
            i -= 1
        array[i] = n
    return array
 
array = insertion_sort(x)
print (array)
```

### Пирамидальная сортировка

```python
def heap_sort_helper(A, n, i):
    l=left(i)
    r=right(i)
    if l<n and A[l]>A[i]:
        largest = 1
    else:
        largest = i
    if r<n and A[r]> A[largest]:
        largest=r
    if largest !=i:
        A[i], A[largest]= A[largest], A[i]
        heap_sort_helper(A, n, largest)
def left(i):
    return 2*i+1
def right(i):
    return 2*i+2
def heap_sort(A):
    n=len(A)
    for i in range(n,-1,-1):
        heap_sort_helper(A,n,i)
    for i in range(n-1, 0, -1):
        heap_sort_helper(A, i, 0)
a=[x]
heap_sort(a)
print(a)
```

### Быстрая сортировка

```python
def quick_sort(array):
    quik_sort_helper(array, 0, len(array)-1)
    return array
def quik_sort_helper(array, low, high):
    if low< high:
        split_point = partition(array, low, high)
        quik_sort_helper(array, low, split_point-1)
        quik_sort_helper(array, split_point+1, high)  
def partition(array, low, high):
    pivot_value=array(low)
    left_mark=low+1
    right_mark=high
    done=False
    while not done:
        while left_mark<= right_mark and array[left_mark]<= pivot_value:
            left_mark+=1
        while right_mark >= left_mark and array[left_mark]>= pivot_value:
            right_mark-=1
        if right_mark < left_mark:
            done =True
        else:
            array[left_mark], array[right_mark]= array[right_mark], array[left_mark]
    array[low], array[right_mark] = array[right_mark], array[low]
    return right_mark
array=quick_sort([x])
print(array)
```

### Сортировка Шелла

```python
def shell_sort(array):
    n= len(array)
    step=n//2
    while step>0:
        for i in range (step, n):
            j=i
            while j>= step and array[j]<array[j-step]:
                array[j], array[j-step] =array[j-step],array[j]
                j=j-step
                step= step//2
                
arr=shell_sort(x)
print(arr)
```
